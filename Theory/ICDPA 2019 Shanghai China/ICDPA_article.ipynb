```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
LOG_FILE = 'C:/_Repositories/Building_Detection/unet.log'
```

```python
def get_jaccard_from_log(log_file):
    with open (log_file, 'r') as f:
        log = f.readlines()
    dscs = [float(s.split()[-4]) for s in log if 'val_jaccard_idx' in s]
    return dscs, max(dscs), dscs.index(max(dscs))
```

```python
unet_js, max_j_unet, idx_unet = get_jaccard_from_log(LOG_FILE)
```

```python
epochs = range(1, 101)
plt.figure(figsize=(17, 13))
plt.rcParams.update({'font.size': 24})
plt.xlabel('E')
plt.ylabel('J')
plt.plot(epochs, unet_js, color = 'darkblue', label='U-Net', linewidth=3)
plt.legend()
plt.grid()
plt.savefig('js.png')
```
